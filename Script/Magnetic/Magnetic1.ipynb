```python
# -*- coding: utf-8 -*-
"""
Magnetic Response plot (datasheet style): OUT vs B
Input CSV: /content/Lab_Magnetic.csv

Required columns (Thai headers in your file):
ระยะ (mm) | ขั้ว | Shield | Vout (mV) | ค่า ADC (counts) | ΔV (V) | VQ (mV) |
ΔV = Vout – VQ (V) | Sensitivity S25°C (mV/mT) | Stc | Ta | B (mT)
"""
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ---------- Config ----------
CSV_PATH = "/content/Lab_Magnetic.csv"
VCC = 3.3
VL_MIN = 0.2
VL_MAX = VCC - 0.2   # 3.1 V
VQ = VCC / 2.0       # 1.65 V
# ---------------------------

# Load
df = pd.read_csv(CSV_PATH)

# Rename to English-friendly keys
col_map = {
    "ระยะ (mm)": "distance_mm",
    "ขั้ว": "pole",
    "Shield": "shield",
    "Vout (mV)": "vout_mV",
    "ค่า ADC (counts)": "adc_counts",
    "ΔV (V)": "dV_V",
    "VQ (mV)": "VQ_mV",
    "ΔV = Vout – VQ (V)": "dV_from_cols_V",
    "Sensitivity S25°C (mV/mT)": "S25_mV_per_mT",
    "Stc": "stc",
    "Ta": "Ta_C",
    "B (mT)": "B_mT",
}
df = df.rename(columns=col_map)

# Parse numerics
for c in ["distance_mm","vout_mV","adc_counts","dV_V","VQ_mV","dV_from_cols_V","S25_mV_per_mT","stc","Ta_C","B_mT"]:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

# Determine Vout units (your table often stores it already in volts)
v = df["vout_mV"].astype(float)
if v.max() is not None and v.max() <= 6.0:
    df["vout_V"] = v
else:
    df["vout_V"] = v / 1000.0

# Keep rows with finite values
df = df[np.isfinite(df["B_mT"]) & np.isfinite(df["vout_V"])]

# Fit linear response only inside the guaranteed linear output window
fit_mask = (df["vout_V"] >= VL_MIN) & (df["vout_V"] <= VL_MAX)
fit_df = df[fit_mask]

if len(fit_df) >= 2:
    a, b = np.polyfit(fit_df["B_mT"].values, fit_df["vout_V"].values, 1)  # Vout = a*B + b
else:
    # Fallback small slope around VQ if insufficient points
    a, b = (0.001, VQ)

# Smooth curve across observed B range
Bx = np.linspace(df["B_mT"].min(), df["B_mT"].max(), 400)
Vfit = a * Bx + b
Vclip = np.clip(Vfit, VL_MIN, VL_MAX)  # emulate saturation at rails

# ===== Plot (one figure, no Thai on-plot text) =====
plt.figure(figsize=(7.5, 6.0))

# Response curve (clipped to rails)
plt.plot(Bx, Vclip, linewidth=2)

# Raw points
plt.scatter(df["B_mT"], df["vout_V"], s=18, alpha=0.75)

# Axes like the datasheet
plt.axvline(0, linewidth=1.2)             # vertical axis at B=0
plt.axhline(0, linewidth=1.0)             # baseline at 0 V

# VCC/2, rails
plt.axhline(VQ, linestyle=":", linewidth=1.0)
plt.text(Bx.min(), VQ + 0.02, "VCC/2", va="bottom")

plt.axhline(VL_MIN, linestyle="--", linewidth=1.0)
plt.text(Bx.min(), VL_MIN + 0.02, "VL (MIN)", va="bottom")

plt.axhline(VL_MAX, linestyle="--", linewidth=1.0)
plt.text(Bx.min(), VL_MAX + 0.02, "VL (MAX)", va="bottom")

# Labels and title (English only)
plt.xlabel("B (mT)   \u2190 north       0 mT        south \u2192")
plt.ylabel("OUT (V)")
plt.title("Magnetic Response (DRV5055, 25°C)")

plt.grid(True, linestyle=":")
plt.tight_layout()
plt.show()
```

```python
# --- DRV5055: compute %attenuation with Shield and plot ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

CSV_PATH = "/content/Lab_Magnetic.csv"

# 1) Load + normalize column names
df = pd.read_csv(CSV_PATH)

col_map = {
    "ระยะ (mm)": "distance_mm",
    "ขั้ว": "pole",                      # "N" / "S"
    "Shield": "shield",                  # "Yes" / "No"
    "B (mT)": "B_mT",
    "Vout (mV)": "vout_mV", # Corrected to vout_mV
    "ค่า ADC (counts)": "adc_counts",
}
df = df.rename(columns=col_map)

# Ensure numeric
for c in ["distance_mm", "B_mT", "vout_mV", "adc_counts"]: # Corrected to vout_mV
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

# Add vout_V column as in previous cells
if "vout_mV" in df.columns:
    v = df["vout_mV"].astype(float)
    df["vout_V"] = v if v.max() <= 6.0 else v / 1000.0
else:
    raise ValueError("Column 'Vout (mV)' not found.")


# 2) Pair rows (No vs Yes) by distance & pole
# Keep only rows that exist for both conditions
no_shield = df[df["shield"].str.lower().eq("no")]
yes_shield = df[df["shield"].str.lower().eq("yes")]

key = ["distance_mm", "pole"]

merged = pd.merge(
    no_shield[key + ["B_mT", "vout_V"]].rename( # Corrected to use vout_V
        columns={"B_mT":"B_no", "vout_V":"V_no"} # Corrected to use V_no
    ),
    yes_shield[key + ["B_mT", "vout_V"]].rename( # Corrected to use vout_V
        columns={"B_mT":"B_yes", "vout_V":"V_yes"} # Corrected to use V_yes
    ),
    on=key, how="inner"
)


# 3) Compute %attenuation (use absolute values for B to ignore polarity)
def pct_atten(a_no, a_yes):
    a_no = np.abs(a_no)
    a_yes = np.abs(a_yes)
    with np.errstate(divide="ignore", invalid="ignore"):
        out = (a_no - a_yes) / a_no * 100.0
    return np.where(np.isfinite(out), out, np.nan)

merged["atten_B_%"]    = pct_atten(merged["B_no"],   merged["B_yes"])
merged["atten_V_%"]    = pct_atten(merged["V_no"],   merged["V_yes"]) # Corrected to use V_no and V_yes

# 4) Save summary table
summary_cols = [
    "distance_mm", "pole",
    "B_no","B_yes","atten_B_%",
    "V_no","V_yes","atten_V_%", # Corrected to use V_no and V_yes
]
summary = merged[summary_cols].sort_values(["pole","distance_mm"])
summary.to_csv("/content/attenuation_summary.csv", index=False)
print("Saved: /content/attenuation_summary.csv")

# 5) Plot %attenuation vs distance (B) for N and S
plt.figure(figsize=(7,6))
for pole in ["N","S"]:
    sub = summary[summary["pole"]==pole]
    if len(sub):
        plt.plot(sub["distance_mm"], sub["atten_B_%"], marker="o", label=f"{pole} (B)")
plt.xlabel("Distance (mm)")
plt.ylabel("% Attenuation (|B|)")
plt.title("Effect of Shield on Magnetic Flux Density")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("/content/fig_attenuation_B_vs_distance.png", dpi=150)
plt.show()

# 6) (Optional) Plot attenuation from voltage as cross-check
plt.figure(figsize=(7,6))
for pole in ["N","S"]:
    sub = summary[summary["pole"]==pole]
    if len(sub):
        plt.plot(sub["distance_mm"], sub["atten_V_%"], marker="o", label=f"{pole} (Vout)") # Corrected to use atten_V_%
plt.xlabel("Distance (mm)")
plt.ylabel("% Attenuation (Vout)")
plt.title("Effect of Shield on Output Voltage")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("/content/fig_attenuation_V_vs_distance.png", dpi=150)
plt.show()
```

```python
# -*- coding: utf-8 -*-
"""
Plot B (mT) as output vs Vout (V) as input.
- Input CSV: /content/Lab_Magnetic.csv
- Figures:
  1) B vs Vout (all data), grouped by pole & shield, with linear fits and R^2
  2) B vs Vout (restricted to linear Vout window ~0.2–3.1 V), with linear fits and R^2
- Console prints linear equations: B = m*Vout + c
"""
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

CSV_PATH = "/content/Lab_Magnetic.csv"

# ---- Load & rename columns ----
df = pd.read_csv(CSV_PATH)
col_map = {
    "ระยะ (mm)": "distance_mm",
    "ขั้ว": "pole",
    "Shield": "shield",
    "Vout (mV)": "vout_mV",
    "ค่า ADC (counts)": "adc_counts",
    "ΔV (V)": "dV_V",
    "VQ (mV)": "VQ_mV",
    "ΔV = Vout – VQ (V)": "dV_from_cols_V",
    "Sensitivity S25°C (mV/mT)": "S25_mV_per_mT",
    "Stc": "stc",
    "Ta": "Ta_C",
    "B (mT)": "B_mT",
    "Vout (V)": "vout_V_direct",
    "Vout ADC (V)": "vout_adc_V",
}
df = df.rename(columns=col_map)

# Determine Vout in volts (prefer direct V column; else convert mV if needed)
if "vout_V_direct" in df.columns and df["vout_V_direct"].notna().any():
    df["Vout_V"] = pd.to_numeric(df["vout_V_direct"], errors="coerce")
else:
    v = pd.to_numeric(df.get("vout_mV", pd.Series(dtype=float)), errors="coerce")
    df["Vout_V"] = v.where(v <= 6.0, v / 1000.0)

# Use provided B (mT)
df["B_mT"] = pd.to_numeric(df["B_mT"], errors="coerce")

# Keep finite rows
df = df[np.isfinite(df["Vout_V"]) & np.isfinite(df["B_mT"]) & df["pole"].notna() & df["shield"].notna()]

# Helper: per-case iterator
def iter_cases(frame):
    for (pole, shield), sub in frame.groupby(["pole", "shield"]):
        sub = sub.sort_values("Vout_V")
        yield f"{pole} - Shield {shield}", sub

# Helper: compute linear fit (B = m*V + c) and R^2
def fit_and_r2(x, y):
    if len(x) < 2:
        return np.nan, np.nan, np.nan
    m, c = np.polyfit(x, y, 1)
    y_pred = m * x + c
    ss_res = np.sum((y - y_pred) ** 2)
    ss_tot = np.sum((y - np.mean(y)) ** 2)
    r2 = 1 - ss_res / ss_tot if ss_tot != 0 else np.nan
    return m, c, r2

# ---------------- Figure 1: ALL DATA ----------------
plt.figure(figsize=(7.5, 6))
for label, sub in iter_cases(df):
    x = sub["Vout_V"].to_numpy(float)
    y = sub["B_mT"].to_numpy(float)
    plt.scatter(x, y, label=f"{label} (data)")
    m, c, r2 = fit_and_r2(x, y)
    if np.isfinite(m):
        xfit = np.linspace(x.min(), x.max(), 100)
        yfit = m * xfit + c
        plt.plot(xfit, yfit, linestyle="--", label=f"{label} (fit)")

plt.xlabel("Vout (V)")
plt.ylabel("B (mT)")
plt.title("Figure 1: B vs Vout (all data)")
plt.grid(True)
plt.legend(ncol=2)
plt.tight_layout()
plt.show()

# Print equations for all data
print("\nLinear equations on ALL data (B = m*Vout + c):")
for label, sub in iter_cases(df):
    x = sub["Vout_V"].to_numpy(float)
    y = sub["B_mT"].to_numpy(float)
    m, c, r2 = fit_and_r2(x, y)
    print(f"{label}: B = {m:.4f} * Vout + {c:.4f}, R^2 = {r2:.4f}")

# ---------------- Figure 2: RESTRICTED TO LINEAR WINDOW ----------------
VL_MIN, VL_MAX = 0.2, 3.1
mask = (df["Vout_V"] >= VL_MIN) & (df["Vout_V"] <= VL_MAX)
df_lin = df[mask].copy()

plt.figure(figsize=(7.5, 6))
for label, sub in iter_cases(df_lin):
    x = sub["Vout_V"].to_numpy(float)
    y = sub["B_mT"].to_numpy(float)
    plt.scatter(x, y, label=f"{label} (data)")
    m, c, r2 = fit_and_r2(x, y)
    if np.isfinite(m):
        xfit = np.linspace(max(x.min(), VL_MIN), min(x.max(), VL_MAX), 100)
        yfit = m * xfit + c
        plt.plot(xfit, yfit, linestyle="--", label=f"{label} (fit)")

# Decoration
plt.axvline(0.2, linestyle=":", linewidth=1)
plt.axvline(3.1, linestyle=":", linewidth=1)
plt.text(0.22, plt.gca().get_ylim()[1]*0.9, "Linear window", va="top")
plt.xlabel("Vout (V)")
plt.ylabel("B (mT)")
plt.title("Figure 2: B vs Vout (linear window 0.2–3.1 V)")
plt.grid(True)
plt.legend(ncol=2)
plt.tight_layout()
plt.show()

# Print equations for linear-window data
print("\nLinear equations in linear Vout window (B = m*Vout + c):")
for label, sub in iter_cases(df_lin):
    x = sub["Vout_V"].to_numpy(float)
    y = sub["B_mT"].to_numpy(float)
    m, c, r2 = fit_and_r2(x, y)
    print(f"{label}: B = {m:.4f} * Vout + {c:.4f}, R^2 = {r2:.4f}")
```

```python
# -*- coding: utf-8 -*-
"""
Compare real measured B vs B predicted by linear fit
for DRV5055 Magnetic Sensor (N/S poles, with & without shield)
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ====== Data ======
data = {
    "distance_mm": [1.000,1.500,2.000,2.500,3.000,3.500,4.000,4.500,
                    1.000,1.500,2.000,2.500,3.000,3.500,4.000,4.500],
    "pole": ["N","N","N","N","N","N","N","N",
             "S","S","S","S","S","S","S","S"],
    "B_no": [27.550,27.550,12.767,5.900,3.050,1.917,1.300,0.883,
             -25.633,-25.433,-13.450,-5.800,-2.800,-1.600,-1.600,-1.017],
    "B_yes": [27.533,27.517,11.683,5.483,3.017,1.817,1.233,0.867,
              -25.533,-25.367,-11.150,-4.467,-2.317,-1.250,-0.717,-0.417],
}
df = pd.DataFrame(data)

# ====== Linear fit ======
fits = []
for pole in ["N", "S"]:
    for shield_label, col in zip(["No", "Yes"], ["B_no", "B_yes"]):
        sub = df[df["pole"] == pole]
        x = sub["distance_mm"]
        y = sub[col]
        m, c = np.polyfit(x, y, 1)
        R2 = 1 - np.sum((y - (m*x + c))**2) / np.sum((y - np.mean(y))**2)
        fits.append((pole, shield_label, m, c, R2))

# ====== Plot ======
plt.figure(figsize=(8,6))
for pole, color in zip(["N","S"], ["tab:blue","tab:red"]):
    sub = df[df["pole"] == pole]
    for shield_label, linestyle, alpha in zip(["No","Yes"], ["-", "--"], [1.0, 0.9]):
        col = f"B_{shield_label.lower()}"
        m, c, R2 = [(m, c, R2) for p,s,m,c,R2 in fits if p==pole and s==shield_label][0]
        xfit = np.linspace(sub["distance_mm"].min(), sub["distance_mm"].max(), 100)
        yfit = m * xfit + c

        # Plot data points
        plt.scatter(sub["distance_mm"], sub[col], color=color, alpha=0.8,
                    label=f"{pole} (Shield {shield_label}) data")
        # Plot linear fit
        plt.plot(xfit, yfit, linestyle=linestyle, color=color, alpha=alpha,
                 label=f"{pole} (Shield {shield_label}) fit, R²={R2:.3f}")

plt.axhline(0, color="gray", linestyle=":")
plt.xlabel("Distance (mm)")
plt.ylabel("Magnetic Flux Density B (mT)")
plt.title("Comparison: Real B vs Linear Predicted B (by Shield & Pole)")
plt.grid(True, linestyle=":")
plt.legend(ncol=2, fontsize=9)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ====== Data ======
data = {
    "distance_mm": [1.000,1.500,2.000,2.500,3.000,3.500,4.000,4.500,
                    1.000,1.500,2.000,2.500,3.000,3.500,4.000,4.500],
    "pole": ["N","N","N","N","N","N","N","N",
             "S","S","S","S","S","S","S","S"],
    "B_no": [27.550,27.550,12.767,5.900,3.050,1.917,1.300,0.883,
             -25.633,-25.433,-13.450,-5.800,-2.800,-1.600,-1.600,-1.017],
    "B_yes": [27.533,27.517,11.683,5.483,3.017,1.817,1.233,0.867,
              -25.533,-25.367,-11.150,-4.467,-2.317,-1.250,-0.717,-0.417],
}

df = pd.DataFrame(data)

# ====== Fit linear models ======
fits = []
for pole in ["N", "S"]:
    for shield_label, col in zip(["No", "Yes"], ["B_no", "B_yes"]):
        sub = df[df["pole"] == pole]
        x = sub["distance_mm"].astype(float)
        y = sub[col].astype(float)
        m, c = np.polyfit(x, y, 1)
        R2 = 1 - np.sum((y - (m*x + c))**2) / np.sum((y - np.mean(y))**2)
        fits.append((pole, shield_label, m, c, R2))

# ====== Plot ======
plt.figure(figsize=(8,6))

for pole, color in zip(["N","S"], ["tab:blue","tab:red"]):
    sub = df[df["pole"] == pole]
    for shield_label, linestyle in zip(["No","Yes"], ["-", "--"]):
        col = f"B_{shield_label.lower()}"
        m, c, R2 = [(m, c, R2) for p,s,m,c,R2 in fits if p==pole and s==shield_label][0]
        xfit = np.linspace(sub["distance_mm"].min(), sub["distance_mm"].max(), 100)
        yfit = m * xfit + c

        # Real data
        plt.scatter(sub["distance_mm"], sub[col], color=color, alpha=0.8,
                    label=f"{pole} Shield {shield_label} (data)")
        # Linear fit
        plt.plot(xfit, yfit, linestyle=linestyle, color=color,
                 label=f"{pole} Shield {shield_label} (fit, R²={R2:.3f})")

plt.axhline(0, color="gray", linestyle=":")
plt.xlabel("Distance (mm)")
plt.ylabel("Magnetic Flux Density B (mT)")
plt.title("Measured vs Linear Predicted Magnetic Field (B vs Distance)")
plt.grid(True, linestyle=":")
plt.legend(ncol=2, fontsize=9)
plt.tight_layout()
plt.show()
```
